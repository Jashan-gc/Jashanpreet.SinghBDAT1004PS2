# Question 3
Write a class named Marsupial that can be used as shown below:  
> m = Marsupial()  
> m.put_in_pouch('doll')  
> m.put_in_pouch('firetruck')  
> m.put_in_pouch('kitten')  
> m.pouch_contents()  
['doll', 'firetruck', 'kitten']  
  

  Now write a class named Kangaroo as a subclass of Marsupial that inherits all the  
attributes of Marsupial and also:  
a. extends the Marsupial __init__ constructor to take, as input, the  
coordinates x and y of the Kangaroo object,  
b. supports method jump that takes number values dx and dy as input and  
moves the kangaroo by dx units along the x-axis and by dy units along the y-  
axis, and  
c. overloads the __str__ operator so it behaves as shown below.  
> k = Kangaroo(0,0)  
> print(k)  
I am a Kangaroo located at coordinates (0,0)  
> k.put_in_pouch('doll')   
> k.put_in_pouch('firetruck')  
> k.put_in_pouch('kitten')  
> k.pouch_contents()  

['doll', 'firetruck', 'kitten']

> k.jump(1,0)  
> k.jump(1,0)  
> k.jump(1,0)  
> print(k)  

I am a Kangaroo located at coordinates (3,0)  

```python
class Marsupial:
    def __init__(self):
        self.pouch = []  # Initialize empty list
    
    def put_in_pouch(self, contents):
        self.pouch.append(contents)  # Adds contents to list
    
    def pouch_contents(self):
        print(self.pouch)
        return self.pouch  # Returns the list
    
class Kangaroo(Marsupial):
    def __init__(self, x, y):
        super().__init__() #super class constructor
        self.x = x
        self.y = y
    
    def jump(self, dx, dy): #modifies x and y
        self.x = dx + self.x
        self.y = dy + self.y
    
    def __str__(self): #string representation of object
        return f"I am a Kangaroo located at coordinates ({self.x},{self.y})"


m = Marsupial()  # Create object of class Marsupial
m.put_in_pouch('doll')
m.put_in_pouch('firetruck')
m.put_in_pouch('kitten')
m.pouch_contents()

k = Kangaroo(0, 0)
print(k)
k.put_in_pouch('doll2')
k.put_in_pouch('firetruck2')
k.put_in_pouch('kitten2')
k.pouch_contents()
k.jump(1,0)
k.jump(1,0)
k.jump(1,0)
print(k)
```

# Question 5
Write a recursive method binary() that takes a non-negative  
integer n and prints the binary representation of integer n.  
> binary(0)    
0    
> binary(1)  
1  
> binary(3)  
11  
> binary(9)  
1001  

```python
def binary(n):
    if n<2:
        print(n,end='')
    else:
        binary(n//2) 
        print(n % 2, end='')
        
binary(0)
print()
binary(1)
print()
binary(3)
print()
binary(9)
```

# Question 6  
Implement a class named HeadingParser that can be used to parse an HTML  
document, and retrieve and print all the headings in the document. You should  
implement your class as a subclass of HTMLParser, defined in Standard Library  
module html.parser. When fed a string containing HTML code, your class should  
print the headings, one per line and in the order in which they appear in the  
document. Each heading should be indented as follows: an h1 heading should have    
indentation 0, and h2 heading should have indentation 1, etc. Test your  
implementation using w3c.html.  
> infile = open('w3c.html')  
> content = infile.read()  
> infile.close()  
> hp = HeadingParser()  
> hp.feed(content)  
W3C Mission  
Principles  

```python
from html.parser import HTMLParser

class HeadingParser(HTMLParser):
    def __init__(self):
        super().__init__()  # super class constructor
        self.headings_list = []  # initialization of an empty list and indent variable
        self.indent = 0
        self.is_title = False #code to exclude title as heading tag

    def handle_starttag(self, tag, attrs):
        if tag.startswith('h') and len(tag) == 2:  # checks if the tag is a heading tag
            self.indent = int(tag[1]) - 1  # e.g. making indent 0 if tag is h1
        if tag == 'title':
            self.is_title = True
            
    def handle_endtag(self, tag):
        if tag == 'title':
            self.is_title = False

    def handle_data(self, data):
        heading = data.strip()  # remove white spaces
        if heading and not self.is_title: #
            self.headings_list.append(' ' * self.indent + heading) #adds whitespace according to indentation and appends heaing names to the list
        
    def print_data(self):
        for heading in self.headings_list:
            print(heading)

infile = open('w3c.txt')
content = infile.read()
infile.close()

hp = HeadingParser()
hp.feed(content)
hp.print_data()
```

# Question 7
Implement recursive function webdir() that takes as input: a URL (as a string) and  
non-negative integers depth and indent. Your function should visit every web  
page reachable from the starting URL web page in depth clicks or less, and print  
each web page's URL. As shown below, indentation, specified by indent, should  
be used to indicate the depth of a URL.  
>webdir('http://reed.cs.depaul.edu/lperkovic/csc242/test1.html'
, 2, 0)  
http://reed.cs.depaul.edu/lperkovic/csc242/test1.html  
http://reed.cs.depaul.edu/lperkovic/csc242/test2.html  
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html  
http://reed.cs.depaul.edu/lperkovic/csc242/test3.html  
http://reed.cs.depaul.edu/lperkovic/csc242/test4.html  

```python
import re
import requests

def webdir(url, depth, indent=0):
    print(' ' * indent + url) # Prints the URL with relevant indent

    if depth == 0: #if depth is 0, then returns 0
        return 0

    try:
        response = requests.get(url)

        urls = re.findall(r'href=[\'"]?([^\'" >]+)', response.text) #extracts all URLs with a regular expression(for hyperlink) and stores in list urls
        for extracted_url in urls:
            if not extracted_url or extracted_url.startswith('#'):   # skips empty URLs or URLs starting with '#'
                continue
            if not extracted_url.startswith('http'): #relative to absolute url
                extracted_url = url + extracted_url

            webdir(extracted_url, depth - 1, indent + 2) #recursive function call on decreased depth and increased indentation

    except requests.exceptions.RequestException as e:
        print('Error:', e)  #Handles errors by printing error messages

# Example usage:
webdir('https://www.rollingstone.com/tv-movies/tv-movie-news/space-jam-forever-the-website-that-wouldnt-die-70507/', 1,0)
```
